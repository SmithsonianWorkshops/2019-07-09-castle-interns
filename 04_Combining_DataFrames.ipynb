# Combining DataFrames
This **notebook** will demonstrate how to combine *DataFrames*

```python
import pandas as pd
```

```python
surveys_df = pd.read_csv('https://ndownloader.figshare.com/files/10717177')
```

```python
top6 = surveys_df.head(6)
top6.info()
```

```python
tail20 = surveys_df.tail(20)
```

```python
combined_df = pd.concat([top6, tail20])
combined_df.info()
```

```python
horizontal = pd.concat([top6, tail20], axis='columns')
horizontal.head()
```

```python
combined_df.describe()
```

```python
combined_df.to_csv('combined.tsv',sep='\t',index=False)
```

```python
combined_df.head()
```

```python
species_df = pd.read_csv('https://ndownloader.figshare.com/files/3299483')
species_df.head()
```

```python
joined_df = pd.merge(left=surveys_df, right=species_df, on='species_id')
joined_df.head()
```

```python
species_df.head(20)
```

```python
joined_df.info()
```

```python
merged_left = pd.merge(left=surveys_df, right=species_df, on='species_id', how='left')
merged_left.info()
```

```python
taxaplot = joined_df.groupby('plot_id')['taxa'].count()
%matplotlib inline
taxaplot.plot(kind= 'bar')
```
